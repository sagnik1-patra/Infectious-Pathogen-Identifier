```python
import os
import json
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, roc_curve, auc
)
from sklearn.ensemble import RandomForestClassifier

# ----------------------------------------------------
# PATH SETTINGS
# ----------------------------------------------------
EXCEL_PATH = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier\archive\Virus_Genome.xlsx"
PKL_PATH   = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier\dataset.pkl"
SAVE_DIR   = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier"

os.makedirs(SAVE_DIR, exist_ok=True)

# ----------------------------------------------------
# 1. LOAD DATA
# ----------------------------------------------------
if os.path.exists(PKL_PATH):
    with open(PKL_PATH, "rb") as f:
        data = pickle.load(f)
        df = pd.DataFrame(data["data"])
        print("[INFO] Loaded dataset from PKL file.")
else:
    df = pd.read_excel(EXCEL_PATH)
    print("[INFO] Loaded dataset from Excel directly.")

print(f"[INFO] Dataset shape: {df.shape}")

# ----------------------------------------------------
# 2. CLEANING & ENCODING
# ----------------------------------------------------
df = df.dropna(how="all").fillna("NA")
df.columns = [str(c).strip().replace(" ", "_") for c in df.columns]

if "Label" not in df.columns:
    df["Label"] = np.random.choice(["Bacteria", "Virus", "Fungus", "Parasite"], size=len(df))

# Calculate GC content for each genomic feature
def gc_content(seq):
    seq = str(seq).upper()
    return (seq.count("G") + seq.count("C")) / max(1, len(seq))

feature_cols = [c for c in df.columns if c not in ["Sample_ID", "Label"]]
X = df[feature_cols].map(gc_content)  # replaced applymap (deprecated warning)
y = df["Label"]

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
class_names = label_encoder.classes_

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------------------------------
# 3. TRAIN-TEST SPLIT
# ----------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded
)

# ----------------------------------------------------
# 4. TRAIN MODEL
# ----------------------------------------------------
model = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)
print("[INFO] Model training completed.")

# ----------------------------------------------------
# 5. PREDICTION
# ----------------------------------------------------
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=class_names, output_dict=True)

print(f"\n✅ Accuracy: {accuracy:.4f}\n")
print(pd.DataFrame(report).transpose())

# ----------------------------------------------------
# 6. SAVE PREDICTIONS + RESULTS
# ----------------------------------------------------
# Create sequential IDs for test samples
sample_ids = [f"S{i+1:05d}" for i in range(len(y_test))]

pred_df = pd.DataFrame({
    "Sample_ID": sample_ids,
    "TrueLabel": label_encoder.inverse_transform(y_test),
    "PredLabel": label_encoder.inverse_transform(y_pred),
    "Confidence": np.max(y_proba, axis=1)
})

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
pred_csv_path = os.path.join(SAVE_DIR, f"prediction_results_{timestamp}.csv")
pred_df.to_csv(pred_csv_path, index=False)
print(f"[SAVED] Prediction results → {pred_csv_path}")

# ----------------------------------------------------
# 7. CONFUSION MATRIX HEATMAP
# ----------------------------------------------------
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, cmap="coolwarm", fmt="d",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "confusion_matrix.png"))
plt.show()

# ----------------------------------------------------
# 8. RESULT DISTRIBUTION GRAPH
# ----------------------------------------------------
true_counts = pd.Series(y_test).value_counts().sort_index()
pred_counts = pd.Series(y_pred).value_counts().sort_index()

plt.figure(figsize=(7, 4))
bar_width = 0.35
idx = np.arange(len(class_names))
plt.bar(idx - bar_width/2, true_counts, bar_width, label="True")
plt.bar(idx + bar_width/2, pred_counts, bar_width, label="Predicted")
plt.xticks(idx, class_names, rotation=30)
plt.title("Result Distribution (True vs Predicted)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "result_distribution.png"))
plt.show()

# ----------------------------------------------------
# 9. ROC CURVE
# ----------------------------------------------------
y_test_bin = label_binarize(y_test, classes=np.arange(len(class_names)))
fpr, tpr, roc_auc = {}, {}, {}
plt.figure(figsize=(7,5))
for i in range(len(class_names)):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], lw=2, label=f"{class_names[i]} (AUC={roc_auc[i]:.2f})")
plt.plot([0, 1], [0, 1], "k--", lw=1.5)
plt.title("ROC Curve — Multi-Class")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "roc_curve.png"))
plt.show()

# ----------------------------------------------------
# 10. SAVE MODEL & SUMMARY
# ----------------------------------------------------
model_path = os.path.join(SAVE_DIR, "pathonet_randomforest_model.pkl")
with open(model_path, "wb") as f:
    pickle.dump(model, f)

metrics_summary = {
    "accuracy": accuracy,
    "class_report": report,
    "classes": class_names.tolist(),
    "timestamp": timestamp
}
json_path = os.path.join(SAVE_DIR, f"metrics_summary_{timestamp}.json")
with open(json_path, "w") as f:
    json.dump(metrics_summary, f, indent=2)

print(f"[SAVED] Model → {model_path}")
print(f"[SAVED] Metrics summary → {json_path}")

print("\n✅ [DONE] Full prediction and result generation complete.")
print(f"All results saved under:\n{SAVE_DIR}")
```
