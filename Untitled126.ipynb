```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (
    confusion_matrix, accuracy_score, classification_report
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# ----------------------------------------------------
# PATHS
# ----------------------------------------------------
EXCEL_PATH = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier\archive\Virus_Genome.xlsx"
PKL_PATH   = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier\dataset.pkl"
SAVE_DIR   = r"C:\Users\NXTWAVE\Downloads\Infectious Pathogen Identifier"

os.makedirs(SAVE_DIR, exist_ok=True)

# ----------------------------------------------------
# 1. LOAD DATA
# ----------------------------------------------------
if os.path.exists(PKL_PATH):
    with open(PKL_PATH, "rb") as f:
        data = pickle.load(f)
        df = pd.DataFrame(data["data"])
        print("[INFO] Loaded data from PKL file.")
else:
    df = pd.read_excel(EXCEL_PATH)
    print("[INFO] Loaded data from Excel directly.")

print(f"[INFO] Dataset shape: {df.shape}")

# ----------------------------------------------------
# 2. PREPROCESSING
# ----------------------------------------------------
df = df.dropna(how="all")
df = df.fillna("NA")

# if no labels exist, create dummy categorical labels for demonstration
if "Label" not in df.columns:
    df["Label"] = np.random.choice(["Bacteria", "Virus", "Fungus", "Parasite"], size=len(df))

# Encode features: convert DNA-like sequences to simple numerical features (GC content etc.)
def gc_content(seq):
    seq = str(seq).upper()
    g = seq.count("G")
    c = seq.count("C")
    return (g + c) / max(1, len(seq))

feature_cols = [c for c in df.columns if c not in ["Sample_ID", "Label"]]
X = df[feature_cols].applymap(gc_content)
y = LabelEncoder().fit_transform(df["Label"])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------------------------------
# 3. TRAIN / TEST SPLIT
# ----------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)

# ----------------------------------------------------
# 4. TRAIN MULTIPLE MODELS
# ----------------------------------------------------
models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(kernel="rbf", probability=True, random_state=42),
    "NeuralNet": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
}

results = {}
for name, model in models.items():
    print(f"\n[INFO] Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = {"model": model, "acc": acc, "pred": y_pred}
    print(f"[RESULT] {name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

# ----------------------------------------------------
# 5. ACCURACY GRAPH (Comparison)
# ----------------------------------------------------
acc_df = pd.DataFrame({
    "Model": list(results.keys()),
    "Accuracy": [v["acc"] for v in results.values()]
})

plt.figure(figsize=(6,4))
sns.barplot(data=acc_df, x="Model", y="Accuracy", palette="viridis")
plt.title("Model Accuracy Comparison")
plt.ylim(0,1)
for i, v in enumerate(acc_df["Accuracy"]):
    plt.text(i, v + 0.01, f"{v:.2f}", ha="center", fontsize=10)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "accuracy_graph.png"))
plt.show()

# ----------------------------------------------------
# 6. CONFUSION MATRIX HEATMAP (for best model)
# ----------------------------------------------------
best_model_name = acc_df.sort_values("Accuracy", ascending=False).iloc[0]["Model"]
best_model = results[best_model_name]["model"]
best_pred = results[best_model_name]["pred"]

cm = confusion_matrix(y_test, best_pred)
labels = sorted(df["Label"].unique())

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
            xticklabels=labels, yticklabels=labels)
plt.title(f"Confusion Matrix  {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "confusion_heatmap.png"))
plt.show()

# ----------------------------------------------------
# 7. PREDICTION GRAPH (True vs Predicted)
# ----------------------------------------------------
plt.figure(figsize=(7,4))
plt.plot(y_test, "o-", label="True", alpha=0.7)
plt.plot(best_pred, "s-", label="Predicted", alpha=0.7)
plt.title(f"Prediction Graph  {best_model_name}")
plt.xlabel("Sample Index")
plt.ylabel("Class ID")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "prediction_graph.png"))
plt.show()

# ----------------------------------------------------
# 8. RESULT GRAPH (Distribution of predicted classes)
# ----------------------------------------------------
pred_classes = pd.Series(best_pred).value_counts().sort_index()
true_classes = pd.Series(y_test).value_counts().sort_index()

plt.figure(figsize=(7,4))
bar_width = 0.35
idx = np.arange(len(labels))
plt.bar(idx - bar_width/2, true_classes, bar_width, label='True')
plt.bar(idx + bar_width/2, pred_classes, bar_width, label='Predicted')
plt.xticks(idx, labels, rotation=30)
plt.title("Result Distribution  True vs Predicted")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "result_graph.png"))
plt.show()

# ----------------------------------------------------
# 9. COMPARISON GRAPH (Accuracy across models)
# ----------------------------------------------------
plt.figure(figsize=(7,5))
sns.lineplot(data=acc_df, x="Model", y="Accuracy", marker="o", linewidth=2.5)
plt.title("Comparison Graph  Accuracy across Models")
plt.ylim(0, 1)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "comparison_graph.png"))
plt.show()

# ----------------------------------------------------
# 10. SAVE METRICS SUMMARY
# ----------------------------------------------------
summary_path = os.path.join(SAVE_DIR, "model_results_summary.csv")
acc_df.to_csv(summary_path, index=False)
print(f"[SAVED] Model comparison summary  {summary_path}")

print("\n [DONE] Visualization complete.")
print(f"Graphs saved under:\n{SAVE_DIR}")
```
